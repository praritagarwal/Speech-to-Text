### Spectrograms

The first option for an audio feature representation is the [spectrogram](https://www.youtube.com/watch?v=_FatxGN3vAM). The code that we give you returns the spectrogram as a 2D tensor, where the first (_vertical_) dimension indexes time, and the second (_horizontal_) dimension indexes frequency.  To speed the convergence of your algorithm, we have also normalized the spectrogram. ### Mel-Frequency Cepstral Coefficients (MFCCs)

The second option for an audio feature representation is [MFCCs](https://en.wikipedia.org/wiki/Mel-frequency_cepstrum).  You do **not** need to dig deeply into the details of how MFCCs are calculated, but if you would like more information, you are welcome to peruse the [documentation](https://github.com/jameslyons/python_speech_features) of the `python_speech_features` Python package.  Just as with the spectrogram features, the MFCCs are normalized in the supplied code.

The main idea behind MFCC features is the same as spectrogram features: at each time window, the MFCC feature yields a feature vector that characterizes the sound within the window.  Note that the MFCC feature is much lower-dimensional than the spectrogram feature, which could help an acoustic model to avoid overfitting to the training dataset. 

````python
from data_generator import plot_mfcc_feature

# plot normalized MFCC
plot_mfcc_feature(vis_mfcc_feature)
# print shape of MFCC
display(Markdown('**Shape of MFCC** : ' + str(vis_mfcc_feature.shape)))
````

Given their effectiveness in modeling sequential data, the first acoustic model you will use is an RNN.  As shown in the figure below, the RNN we supply to you will take the time sequence of audio features as input.

<img src="images/simple_rnn.png" width="50%">

If you would like to peek under the hood at how characters are mapped to indices in the probability vector, look at the `char_map.py` file in the repository. ### Compare the Models

Execute the code cell below to evaluate the performance of the drafted deep learning models.  The training and validation loss are plotted for each model.

````python
from glob import glob
import numpy as np
import _pickle as pickle
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style(style='white')

# obtain the paths for the saved model history
all_pickles = sorted(glob("results/*.pickle"))
# extract the name of each model
model_names = [item[8:-7] for item in all_pickles]
# extract the loss history for each model
valid_loss = [pickle.load( open( i, "rb" ) )['val_loss'] for i in all_pickles]
train_loss = [pickle.load( open( i, "rb" ) )['loss'] for i in all_pickles]
# save the number of epochs used to train each model
num_epochs = [len(valid_loss[i]) for i in range(len(valid_loss))]

fig = plt.figure(figsize=(16,5))

# plot the training loss vs. epoch for each model
ax1 = fig.add_subplot(121)
for i in range(len(all_pickles)):
    ax1.plot(np.linspace(1, num_epochs[i], num_epochs[i]), 
            train_loss[i], label=model_names[i])
# clean up the plot
ax1.legend()  
ax1.set_xlim([1, max(num_epochs)])
plt.xlabel('Epoch')
plt.ylabel('Training Loss')

# plot the validation loss vs. epoch for each model
ax2 = fig.add_subplot(122)
for i in range(len(all_pickles)):
    ax2.plot(np.linspace(1, num_epochs[i], num_epochs[i]), 
            valid_loss[i], label=model_names[i])
# clean up the plot
ax2.legend()  
ax2.set_xlim([1, max(num_epochs)])
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
plt.show()
````

 - The number of units in the RNN layer has to be exactly 29, neither less nor more. 
  
 - The RNN layer is forced to output probability distribution for the text labels rather than feature vectors
  
These two limitations severely constrain the ability of model 0 to learn useful patterns in the input data, thereby leading to its underperformance. 

We also notice that using a single layer of GRU with followed by a dense layer, as is done in model 1 gives a significant improvement over model 0. We can immediately understand this improvement by refering to the limitations of model 0 mentioned above. Namely, using a Dense layer, allows us to use as many units in the RNN as we want (200 units in the present case). This allows us to come up with a configuration that can learn many more patterns in the time sequences than was possible in model 0.  Secondly, the RNN layer now outputs feature vectors corresponding to each time-step rather than probability distributions for the text labels. Thus model 1 is able to develop a better function that maps the input sequences to the output labels. 

In model 2, we use CNNs to process the input data. Clearly, as is the case with any CNNs, this allows for translation invariance in the data. Thus allowing the model 1 identify speech patterns irrespective of when they occur in the sequences. It is therefore no surprize that upon using CNNs we obtain a significantly lower training loss. However, this model significantly overfits with the validation loss being ~ 148 as compard to a training loss of ~ 63 i.e. validation loss is more than double the training loss. However, notice that had we stopped training after 11 epochs, validation loss would have been much lower i.e ~ 129 with the training loss being ~ 95. This suggest that using model 2 early stopping can give better results. In order to gauge the effect of kernel size in the CNN layer, in model 5 we tried using a kernel size of 5 (as compared to 11 in model 2). We also tried training it with a smaller learning rate. However, model 5 does not show any significant improvement over model 2. 

**Model 3 seems to perform the best**. Clearly, showing that using a stack of RNNs allows the model to learn better speech patterns. This is consistent with the well known fact that using deep neural networks often gives better results. 

Model 4 uses bidirectional RNNs which in principle can learn how the future portion of a sequence  affect its current output. Note that when we speak we already know the entire word that we are trying to vocalize. Therefore it is natural to expect that an RNN which also extracts information from future portions of a speech will perform better. Unfortunately, this does not seem to be the case with the current model which does not show any significant improvement over model 1 which is same in all other aspects expect for the use of Bidirectional RNNs. Nonetheless, we expect that when combined with other architectures such as using CNN layers and stacking of RNN layers, bidirectional RNNs will probably be quite helpful.

## STEP 3: Obtain Predictions

We have written a function for you to decode the predictions of your acoustic model.  To use the function, please execute the code cell below.

````python
import numpy as np
from data_generator import AudioGenerator
from keras import backend as K
from utils import int_sequence_to_text
from IPython.display import Audio

def get_predictions(index, partition, input_to_softmax, model_path):
    """ Print a model's decoded predictions
    Params:
        index (int): The example you would like to visualize
        partition (str): One of 'train' or 'validation'
        input_to_softmax (Model): The acoustic model
        model_path (str): Path to saved acoustic model's weights
    """
    # load the train and test data
    data_gen = AudioGenerator()
    data_gen.load_train_data()
    data_gen.load_validation_data()
    
    # obtain the true transcription and the audio features 
    if partition == 'validation':
        transcr = data_gen.valid_texts[index]
        audio_path = data_gen.valid_audio_paths[index]
        data_point = data_gen.normalize(data_gen.featurize(audio_path))
    elif partition == 'train':
        transcr = data_gen.train_texts[index]
        audio_path = data_gen.train_audio_paths[index]
        data_point = data_gen.normalize(data_gen.featurize(audio_path))
    else:
        raise Exception('Invalid partition!  Must be "train" or "validation"')
        
    # obtain and decode the acoustic model's predictions
    input_to_softmax.load_weights(model_path)
    print('loaded the weights')
    prediction = input_to_softmax.predict(np.expand_dims(data_point, axis=0))
    output_length = [input_to_softmax.output_length(data_point.shape[0])] 
    pred_ints = (K.eval(K.ctc_decode(
                prediction, output_length)[0][0])+1).flatten().tolist()
    
    # play the audio file, and display the true and predicted transcriptions
    print('-'*80)
    Audio(audio_path)
    print('True transcription:\n' + '\n' + transcr)
    print('-'*80)
    print('Predicted transcription:\n' + '\n' + ''.join(int_sequence_to_text(pred_ints)))
    print('-'*80)
````
